
## The function given below provides data retrieval without using the Twitter API. 

### The dataset consists of text, date, view, like, reply, retweet columns.

```python
from bs4 import BeautifulSoup
import sys
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import snscrape.modules.twitter as sntwitter
import pandas as pd
import csv 
```

```python
def veri_cek():
    
    sayfa = int(input("number of scrolls: "))
    
    
    #Add the chrome extension path.
    browser = webdriver.Chrome(r"C:\Users\...\wQuality-Twitter\Chrome\chromedriver.exe")
    browser.get('https://twitter.com/search?f=live&q=(%23WaterPollution%20OR%20%23CleanWater%20OR%20%23SaveWater%20OR%20%23StopWaterPollution%20OR%20%23CleanWaterMatters%20OR%20%23ProtectOurWater%20OR%20%23WaterCrisis%20OR%20%23WaterIsLife%20OR%20%23ProtectCleanWater%20OR%20%23SustainableWater%20OR%20%23SaveOurWater%20OR%20%23WaterQuality%20OR%20%23WaterConservation%20OR%20%23SaveOurLakes%20OR%20%23CleanWaterForAll%20OR%20%23WaterProtection%20OR%20%23GreenInfrastructure%20OR%20%23WaterRights%20OR%20%23WaterSecurity%20OR%20%23WaterSustainability%20OR%20%23WaterProblem%20OR%20%23WaterIssue)%20lang%3Aen%20until%3A2023-03-21%20since%3A2023-01-01&src=typed_query')
    #The link should be created using the twitter advanced search option.
    
    time.sleep(6)
    
    # File Operations
    
    file = open("hashtags01012023-21032023.csv","w",encoding = "utf-8")
    writer = csv.writer(file)
        #writer.writerow(["text_Id","text"])
    writer.writerow(["text","reply","retweet","like","view","date"])
    

    a = 0
    while a < sayfa:

    #scroll iÅŸlemleri 1 kere scroll 
        lastHeight = browser.execute_script("return document.body.scrollHeight")
        i = 0
        while i<1:
            browser.execute_script("window.scrollTo(0 ,document.body.scrollHeight);")
            time.sleep(6)
            newHeight = browser.execute_script("return document.body.scrollHeight")

            if newHeight == lastHeight:
                break
            else:
                lastHeight = newHeight
            i = i + 1

        time.sleep(6)
        
        pageSource = browser.page_source

        soup = BeautifulSoup(pageSource,"html.parser")
        
        #generic div frame
        tweets = soup.find_all("div",class_='css-1dbjc4n r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu')


        tweetler = []
        
        for tweet in tweets:
                
                #tweet text
                target = tweet.find(class_='css-901oao r-18jsvk2 r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0', attrs={"data-testid": "tweetText"})
                
                if target is not None:
                    targett = target.text
                else:
                    targett = ""
       

                comment_count = 0
                retweet_count = 0
                like_count = 0 
                view_count = 0
            
                #tweet comment count
                target2 = tweet.find(class_='css-18t94o4 css-1dbjc4n r-1777fci r-bt1l66 r-1ny4l3l r-bztko3 r-lrvibr', attrs={"data-testid": "reply"})
                
                if target2 is not None:
                    comment_count = target2.get('aria-label').split(' ')[0]
                else:
                    comment_count = 0
                    
                    
                    
                #tweet retweet count
                target3 = tweet.find(class_='css-18t94o4 css-1dbjc4n r-1777fci r-bt1l66 r-1ny4l3l r-bztko3 r-lrvibr', attrs={"data-testid": "retweet"})
                
                if target3 is not None:
                    retweet_count = target3.get('aria-label').split(' ')[0]
                else:
                    retweet_count = 0
                    
                    
                    
                #tweet like count    
                target4 = tweet.find(class_='css-18t94o4 css-1dbjc4n r-1777fci r-bt1l66 r-1ny4l3l r-bztko3 r-lrvibr', attrs={"data-testid": "like"})
                
                if target4 is not None:
                    like_count = target4.get('aria-label').split(' ')[0]
                else:
                    like_count = 0
                    
                    
                    
                #tweet view count
                target5 = tweet.find(class_='css-4rbku5 css-18t94o4 css-1dbjc4n r-1loqt21 r-1777fci r-bt1l66 r-1ny4l3l r-bztko3 r-lrvibr')
                
                if target5 is not None:
                    try:
                        view_count = int(target5.get('aria-label').split(' ')[0])
                        
                    except ValueError:
                        pass
                else:
                    view_count = 0
                    
                    
                    
                #tweet date
                target6 = tweet.find(class_='css-4rbku5 css-18t94o4 css-901oao r-14j79pv r-1loqt21 r-xoduu5 r-1q142lx r-1w6e6rj r-37j5jr r-a023e6 r-16dba41 r-9aw3ui r-rjixqe r-bcqeeo r-3s2u2q r-qvutc0')
                
                if target6 is not None:
                    date = target6.get('aria-label')
                else:
                    date = 0
            
            
                # if tweet text is same then dataframe will not be written
                if tweet not in tweetler:
                    
                    tweetler.append(tweet)
            
                
                    writer.writerow([targett, comment_count, retweet_count, like_count, view_count, date])
           

        a = a + 1      
```
